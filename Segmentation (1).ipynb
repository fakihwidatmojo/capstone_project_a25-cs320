# **DATA CLEANING (Izza Tsamaro Hammidyah)**

```python
# Mengubah tipe data 'InvoiceDate' menjadi tanggal
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

# Menghapus baris yang tidak memiliki  'CustomerID'
df_clean = df.dropna(subset=["CustomerID"])

# Mengubah CustomerID jadi integer agar tidak ada koma (45866.0 -> 45866)
df_clean['CustomerID'] = df_clean['CustomerID'].astype(int)

# Mengambil data mulai tahun 2022
df_clean = df_clean[df_clean['InvoiceDate'].dt.year >= 2022]

# Mengambil transaksi yang sukses (Tidak dikembalikan)
df_clean = df_clean[df_clean['ReturnStatus'] == 'Not Returned']

# Menghapus baris dengan 'Quantity' dan 'UnitPrice' < 0
df_clean = df_clean[df_clean['Quantity'] > 0]
df_clean = df_clean[df_clean['UnitPrice'] > 0]

# Feature Engineering: TotalPrice (untuk fitur Monetary)
df_clean['DiscountClipped'] = df_clean['Discount'].clip(lower=0, upper=1) # Diskon antara 0% dan 100%
df_clean['TotalPrice'] = df_clean['Quantity'] * df_clean['UnitPrice'] * (1 - df_clean['DiscountClipped'])
```

# **Agregasi RFM**

```python
import datetime as dt

# Menentukan tanggal referensi (Snapshot Date)
# Mengambil tanggal transaksi terakhir di dataset, lalu tambah 1 hari
snapshot_date = df_clean['InvoiceDate'].max() + dt.timedelta(days=1)
print(f"Tanggal Snapshot: {snapshot_date}")

# Menghitung RFM
rfm = df_clean.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days, # Recency: Jarak hari dari transaksi terakhir
    'InvoiceNo': 'nunique',                                  # Frequency: Jumlah order unik (bukan quantity barang)
    'TotalPrice': 'sum'                                      # Monetary: Total uang yang dibelanjakan
}).reset_index()

# 3. Rename kolom agar mudah dipahami
rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalPrice': 'Monetary'
}, inplace=True)

print("Jumlah customer dalam RFM:", len(rfm))
rfm.head()
```

```python
# Cek Distribusi Data Asli (Visualisasi)
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
sns.histplot(rfm['Recency'], kde=True, ax=ax[0])
sns.histplot(rfm['Frequency'], kde=True, ax=ax[1])
sns.histplot(rfm['Monetary'], kde=True, ax=ax[2])
plt.suptitle('Distribusi Data RFM (Sebelum Transformasi)', fontsize=16)
plt.show()
```

# **Scaling (Fakih Widatmojo)**

```python
# Scaling dan Standarisasi RFM
import numpy as np
from sklearn.preprocessing import StandardScaler

rfm_features = rfm[['Recency', 'Frequency', 'Monetary']].copy()

# Log Transformation
# Menggunakan np.log1p (log(x+1)) untuk menghindari error jika ada nilai 0
# Tujuannya agar distribusi data lebih mendekati normal (tidak skewed)
rfm_log = rfm_features.apply(np.log1p)

scaler_rfm = StandardScaler()
rfm_scaled = scaler_rfm.fit_transform(rfm_log)

# Ubah kembali ke DataFrame agar mudah dibaca
rfm_scaled_df = pd.DataFrame(rfm_scaled, columns=['Recency', 'Frequency', 'Monetary'])
rfm_scaled_df.head()

# Cek Distribusi Setelah Transformasi
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
sns.histplot(rfm_scaled_df['Recency'], kde=True, ax=ax[0])
sns.histplot(rfm_scaled_df['Frequency'], kde=True, ax=ax[1])
sns.histplot(rfm_scaled_df['Monetary'], kde=True, ax=ax[2])
plt.suptitle('Distribusi Data RFM (Setelah Log & Scaling)', fontsize=16)
plt.show()
```

# **Clustering KMEANS (Fakih Widatmojo)**

```python
# Cari jumlah cluster terbaik (k) dengan Silhouette Score dan Elbow Method

from sklearn.cluster import KMeans
from sklearn.metrics import (
    silhouette_score,
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report
)

k_values = range(2, 11)

sil_scores = []
inertias = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(rfm_scaled_df)

    sil = silhouette_score(rfm_scaled_df, labels)
    inertia = kmeans.inertia_

    sil_scores.append(sil)
    inertias.append(inertia)

    print(f"k = {k}, Silhouette Score = {sil:.4f}, Inertia = {inertia:.2f}")

# Plot Silhouette & Elbow dalam satu grafik (dua sumbu Y)
fig, ax1 = plt.subplots(figsize=(8, 5))

# Kurva Silhouette
ax1.set_xlabel("Jumlah Cluster (k)")
ax1.set_ylabel("Silhouette Score", color="tab:blue")
ax1.plot(k_values, sil_scores, marker="o", color="tab:blue", label="Silhouette Score")
ax1.tick_params(axis="y", labelcolor="tab:blue")
ax1.grid(True)

# Kurva Elbow (Inertia) di sumbu Y kedua
ax2 = ax1.twinx()
ax2.set_ylabel("Inertia (SSE)", color="tab:red")
ax2.plot(k_values, inertias, marker="s", linestyle="--", color="tab:red", label="Inertia (Elbow)")
ax2.tick_params(axis="y", labelcolor="tab:red")

fig.suptitle("Perbandingan Silhouette Score dan Elbow Method")
fig.tight_layout()
plt.show()
```

```python
# Pilih k terbaik
best_k = 4

kmeans_final = KMeans(n_clusters=best_k, random_state=42, n_init=10)
rfm['Cluster'] = kmeans_final.fit_predict(rfm_scaled_df)

rfm.head()
```

```python
# Statistik tiap cluster
cluster_stats = rfm.groupby('Cluster').agg(
    Recency_mean=('Recency', 'mean'),
    Frequency_mean=('Frequency', 'mean'),
    Monetary_mean=('Monetary', 'mean'),
    Count=('CustomerID', 'count')
).reset_index()

cluster_stats.sort_values('Recency_mean', ascending=False)
```

```python
# Visualisasi Sebaran Cluster (Scatter Plot 3D)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

colors = ['red', 'green', 'blue', 'orange']

for i in range(best_k):
    subset = rfm[rfm['Cluster'] == i]
    ax.scatter(subset['Recency'], subset['Frequency'], subset['Monetary'],
               c=colors[i], label=f'Cluster {i}', s=50, alpha=0.6)

ax.set_xlabel('Recency (Days)')
ax.set_ylabel('Frequency')
ax.set_zlabel('Monetary')
plt.title(f'3D Visualization of {best_k} Customer Segments')
plt.legend()
plt.show()
```

# **Penentuan Kluster Churn**

```python
# Menentukan cluster mana yang dianggap "churn"
# Aturan sederhana:
# - Recency tinggi (lama tidak transaksi)
# - Monetary rendah (nilai belanja kecil)

med_R = rfm['Recency'].median()
med_M = rfm['Monetary'].median()

cluster_stats['IsChurnCluster'] = (
    (cluster_stats['Recency_mean'] >= med_R) &
    (cluster_stats['Monetary_mean'] <= med_M)
)

# Jika tidak ada yang memenuhi, pilih cluster dengan Recency tertinggi & Monetary terendah
if cluster_stats['IsChurnCluster'].sum() == 0:
    # Skor: Recency besar & Monetary kecil â†’ churn
    score = (
        (cluster_stats['Recency_mean'] - cluster_stats['Recency_mean'].min()) /
        (cluster_stats['Recency_mean'].max() - cluster_stats['Recency_mean'].min() + 1e-6)
        -
        (cluster_stats['Monetary_mean'] - cluster_stats['Monetary_mean'].min()) /
        (cluster_stats['Monetary_mean'].max() - cluster_stats['Monetary_mean'].min() + 1e-6)
    )
    chosen = score.idxmax()
    cluster_stats.loc[chosen, 'IsChurnCluster'] = True

cluster_stats
```

```python
# Map cluster -> label churn (1 = churn, 0 = non-churn)
churn_map = cluster_stats.set_index('Cluster')['IsChurnCluster'].to_dict()
churn_map
```

```python
# Menambahkan kolom churn ke rfm
rfm['Churn'] = rfm['Cluster'].map(lambda c: 1 if churn_map[c] else 0)
rfm['Churn'].value_counts()
```

```python
# Agregasi fitur tambahan per customer dari df transaksi
def mode_or_first(x):
    m = x.mode()
    if len(m) > 0:
        return m.iloc[0]
    return x.iloc[0]

cust_agg = df.groupby('CustomerID').agg(
    TotalQuantity=('Quantity', 'sum'),
    AvgUnitPrice=('UnitPrice', 'mean'),
    AvgDiscount=('Discount', 'mean'),
    AvgShippingCost=('ShippingCost', 'mean'),
    Country=('Country', mode_or_first),
    PaymentMethod=('PaymentMethod', mode_or_first),
    Category=('Category', mode_or_first),
    SalesChannel=('SalesChannel', mode_or_first),
    ReturnStatus=('ReturnStatus', mode_or_first),
    ShipmentProvider=('ShipmentProvider', mode_or_first),
    WarehouseLocation=('WarehouseLocation', mode_or_first),
    OrderPriority=('OrderPriority', mode_or_first),
).reset_index()

cust_agg.head()
```

```python
# Digabungkan dengan RFM + label churn
data_cust = pd.merge(rfm, cust_agg, on='CustomerID', how='inner')
# data_cust = data_cust.drop(columns=['Churn']) # Baris ini dihapus agar kolom 'Churn' tetap ada
print(data_cust.shape)
data_cust.head()
```

# **Percobaan Logistic Regression, Random Forest, dan Deep Neural Network**

```python
# One-hot encoding untuk fitur kategorikal
cat_cols = [
    'Country', 'PaymentMethod', 'Category', 'SalesChannel',
    'ReturnStatus', 'ShipmentProvider', 'WarehouseLocation', 'OrderPriority'
]

num_cols = [
    'Recency', 'Frequency', 'Monetary',
    'TotalQuantity', 'AvgUnitPrice', 'AvgDiscount', 'AvgShippingCost'
]

# One-hot encode kolom kategori
data_model = pd.get_dummies(data_cust[num_cols + cat_cols], drop_first=True)

# Target
y = data_cust['Churn'].astype(int)

print("Shape X:", data_model.shape)
print("Distribusi Churn:")
print(y.value_counts(normalize=True))
```

```python
# Train-test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    data_model, y, test_size=0.3, random_state=42, stratify=y
)

print("Train:", X_train.shape, " Test:", X_test.shape)
```

```python
# Standardisasi fitur (supaya cocok untuk Logistic & DNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Logistic Regression & Random Forest
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

models = {}

# Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train_scaled, y_train)
models['LogisticRegression'] = log_reg

# Random Forest
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)  # RF tidak wajib scaled
models['RandomForest'] = rf
```

```python
# Fungsi helper untuk evaluasi

def evaluate_model(name, model, X_tr, X_te, y_tr, y_te, scaled=True):
    if scaled:
        Xtr, Xte = X_tr, X_te
    else:
        Xtr, Xte = X_train, X_test

    y_pred = model.predict(Xte)
    acc = accuracy_score(y_te, y_pred)
    prec = precision_score(y_te, y_pred, zero_division=0)
    rec = recall_score(y_te, y_pred, zero_division=0)
    f1 = f1_score(y_te, y_pred, zero_division=0)

    print(f"=== {name} ===")
    print(f"Accuracy : {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print(f"F1-score : {f1:.4f}")
    print("\nClassification report:")
    print(classification_report(y_te, y_pred, zero_division=0))

    cm = confusion_matrix(y_te, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Evaluasi
evaluate_model("Logistic Regression", log_reg, X_train_scaled, X_test_scaled, y_train, y_test, scaled=True)
evaluate_model("Random Forest", rf, X_train_scaled, X_test_scaled, y_train, y_test, scaled=False)
```

```python
# Membangun arsitektur DNN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

input_dim = X_train_scaled.shape[1]
print("Input dimension:", input_dim)

model_dnn = Sequential([
    Dense(128, activation='relu', input_shape=(input_dim,)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')  # output probabilitas churn
])

model_dnn.compile(
    loss='binary_crossentropy',
    optimizer=Adam(learning_rate=0.001),
    metrics=['accuracy']
)

model_dnn.summary()
```

```python
# Training DNN
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

history = model_dnn.fit(
    X_train_scaled, y_train,
    validation_split=0.2,
    epochs=30,
    batch_size=64,
    callbacks=[early_stop],
    verbose=1
)
```

```python
# Cell 23: Plot loss & accuracy selama training

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.title('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='val')
plt.title('Accuracy')
plt.legend()

plt.show()
```

```python
# Evaluasi DNN di test set
y_pred_proba = model_dnn.predict(X_test_scaled).ravel()
y_pred_dnn = (y_pred_proba >= 0.5).astype(int)

acc = accuracy_score(y_test, y_pred_dnn)
prec = precision_score(y_test, y_pred_dnn, zero_division=0)
rec = recall_score(y_test, y_pred_dnn, zero_division=0)
f1 = f1_score(y_test, y_pred_dnn, zero_division=0)

print("=== DNN ===")
print(f"Accuracy : {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall   : {rec:.4f}")
print(f"F1-score : {f1:.4f}")

print("\nClassification report:")
print(classification_report(y_test, y_pred_dnn, zero_division=0))

cm = confusion_matrix(y_test, y_pred_dnn)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix - DNN')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

Random forest menunjukkan performa terbaik dibandingkan yang lain. Maka dihitung probabilitas Churn menggunakan Random Forest

```python
# Cari kolom churn yang tersedia
if 'churn_label_dnn' in df.columns:
    churn_col = 'churn_label_dnn'
elif 'Churn' in df.columns:
    churn_col = 'Churn'
elif 'churn' in df.columns:
    churn_col = 'churn'
else:
    churn_col = None
    print("âš ï¸ Kolom churn tidak ditemukan, bagian churn_rate akan dikosongkan.")
```

2. Tentukan nama kolom churn yang dipakai

```python
# Pastikan tipe numerik
df['Cluster']  = df['Cluster'].astype(int)
df['Recency']  = df['Recency'].astype(float)
df['Frequency'] = df['Frequency'].astype(float)
df['Monetary'] = df['Monetary'].astype(float)

group_cols = {
    'CustomerID': ('CustomerID', 'nunique'),
    'Recency_mean': ('Recency', 'mean'),
    'Recency_median': ('Recency', 'median'),
    'Frequency_mean': ('Frequency', 'mean'),
    'Monetary_mean': ('Monetary', 'mean'),
    'Monetary_sum': ('Monetary', 'sum'),
}

if churn_col is not None:
    group_cols['Churn_rate'] = (churn_col, 'mean')

cluster_profile = (
    df
    .groupby('Cluster')
    .agg(**group_cols)
    .reset_index()
)

# Persentase customer & share revenue
total_cust = cluster_profile['CustomerID'].sum()
total_rev  = cluster_profile['Monetary_sum'].sum()

cluster_profile['Customer_pct'] = cluster_profile['CustomerID'] / total_cust
cluster_profile['Revenue_share'] = cluster_profile['Monetary_sum'] / total_rev

cluster_profile
```

3. Buat profil cluster (angkaâ€“angka ringkas per segmen)

```python
# Rank: Recency kecil = lebih bagus, Frequency & Monetary besar = lebih bagus
cluster_profile['R_rank'] = cluster_profile['Recency_mean'].rank(ascending=True)
cluster_profile['F_rank'] = cluster_profile['Frequency_mean'].rank(ascending=False)
cluster_profile['M_rank'] = cluster_profile['Monetary_mean'].rank(ascending=False)

cluster_profile['RFM_score'] = (
    cluster_profile['R_rank'] +
    cluster_profile['F_rank'] +
    cluster_profile['M_rank']
)

# Tentukan batas quantile untuk kategorisasi
q1 = cluster_profile['RFM_score'].quantile(1/3)
q2 = cluster_profile['RFM_score'].quantile(2/3)

def label_segment(row):
    score = row['RFM_score']
    churn_rate = row['Churn_rate'] if 'Churn_rate' in row else np.nan

    if score <= q1:
        # Pelanggan terbaik
        if not np.isnan(churn_rate) and churn_rate > 0.4:
            return 'VIP / High Value â€“ At Risk'
        return 'VIP / High Value'
    elif score <= q2:
        # Menengah
        if not np.isnan(churn_rate) and churn_rate > 0.4:
            return 'Loyal â€“ At Risk'
        return 'Loyal / Medium Value'
    else:
        # Nilai rendah
        if not np.isnan(churn_rate) and churn_rate > 0.4:
            return 'Low Value â€“ Almost Lost'
        return 'Low Value'

cluster_profile['SegmentName'] = cluster_profile.apply(label_segment, axis=1)

cluster_profile.sort_values('Cluster')
```

```python
def print_cluster_summary(row):
    c = int(row['Cluster'])
    print(f"=== Cluster {c} â€“ {row['SegmentName']} ===")
    print(f"- Jumlah pelanggan     : {int(row['CustomerID'])} "
          f"({row['Customer_pct']:.1%} dari total)")
    print(f"- Rata-rata Recency    : {row['Recency_mean']:.1f} hari "
          "(semakin kecil = semakin baru bertransaksi)")
    print(f"- Rata-rata Frequency  : {row['Frequency_mean']:.2f} transaksi")
    print(f"- Rata-rata Monetary   : {row['Monetary_mean']:.2f} (rata-rata nilai belanja)")
    print(f"- Total Monetary       : {row['Monetary_sum']:.2f} "
          f"({row['Revenue_share']:.1%} dari total revenue)")

    if 'Churn_rate' in row:
        print(f"- Churn rate (prediksi): {row['Churn_rate']:.1%}")

    # Interpretasi singkat tambahan
    if 'Churn_rate' in row and row['Churn_rate'] > 0.4:
        print("  > Catatan: segmen ini memiliki risiko churn tinggi, "
              "perlu jadi prioritas program retensi.")
    elif 'Churn_rate' in row:
        print("  > Catatan: risiko churn relatif rendah/moderat dibanding segmen lain.")

    print()  # baris kosong


# Cetak semua cluster
for _, row in cluster_profile.sort_values('Cluster').iterrows():
    print_cluster_summary(row)
```
